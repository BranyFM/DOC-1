```python
# Estadísticas Basicas
import numpy as np

# Supongamos ventas mensuales de un producto
ventas = [10000, 12000, 8000, 11000, 9500, 13000, 9000, 10500]

media = np.mean(ventas)
desviacion = np.std(ventas, ddof=1)  # Volatilidad (desviacion estándar)
coef_variacion = (desviacion / media) * 100

print(f"Promedio en ventas: ${media:.2f}")
print(f"desviación estándar (volatilidad): ${desviacion:.2f}")
print(f"Coeficiente de variación (volatilidad relativa): {coef_variacion:.2f}%")
```

```python
# OBTENIENDO INDICADORES DE ESTADISTICA DESCRIPTIVA
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings

warnings.filterwarnings("ignore")

# Definir los datos directamente
datos = [34, 36, 38, 40, 42, 35, 37, 41, 39, 33, 36, 35, 34, 42]
x = np.array(datos)  # Convertir a array de n umpy

# Calcular los estadísticos muestrales
media_aritmetica = np.round(np.mean(x), 2)
moda = stats.mode(x)
moda_valor = moda.mode.item()
mediana = np.round(np.median(x), 2)
cuartiles = np.percentile(x, [25, 50, 75])
primer_cuartil = np.round(cuartiles[0], 2)
tercer_cuartil = np.round(cuartiles[2], 2)
Amplitud_Cuartilica = tercer_cuartil - primer_cuartil
desviacion_estandar_muestral = np.round(np.std(x, ddof=1), 2)
varianza_muestral = np.round(np.var(x, ddof=1), 2)
coef_asimetria = np.round(3 * (media_aritmetica - mediana) / desviacion_estandar_muestral, 2)
coef_variacion_muestral = np.round((desviacion_estandar_muestral / media_aritmetica) * 100, 2)
curtosis = np.round(stats.kurtosis(x), 2)
valor_maximo = np.max(x)
valor_minimo = np.min(x)

# Imprimir los resultados muestrales
print('--- Estadísticos Muestrales ---\n')
print('Número de datos =', len(x))
print('Valor Máximo =', valor_maximo)
print('Valor Mínimo =', valor_minimo)
print('Media Aritmética o Promedio =', media_aritmetica)
print('Moda =', moda_valor)
print('Primer cuartil =', primer_cuartil)
print('Mediana o Segundo Cuartil =', mediana)
print('Tercer cuartil =', tercer_cuartil)
print('Amplitud Cuartilica =', Amplitud_Cuartilica)
print('Desviación estándar muestral =', desviacion_estandar_muestral)
print('Varianza muestral =', varianza_muestral)
print('Coeficiente de Asimetría de Pearson =', coef_asimetria)
print('Coeficiente de Variación muestral en Porcentaje CV =', coef_variacion_muestral)
#print('Curtosis =', curtosis)

# Calcular los estadísticos poblacionales
desviacion_estandar_poblacional = np.round(np.std(x, ddof=0), 2)
varianza_poblacional = np.round(np.var(x, ddof=0), 2)
coef_variacion_poblacional = np.round((desviacion_estandar_poblacional / media_aritmetica) * 100, 2)
asimetria_poblacional = np.round(stats.skew(x, bias=False), 2)
curtosis_poblacional = np.round(stats.kurtosis(x, bias=False), 2)

# Imprimir los resultados poblacionales
print('\n--- Estadísticos Poblacionales ---\n')
print('Desviación estándar poblacional =', desviacion_estandar_poblacional)
print('Varianza poblacional =', varianza_poblacional)
print('Coeficiente de Variación poblacional CV =', coef_variacion_poblacional)
print('Coeficiente de Asimetría poblacional =', coef_asimetria)
print('Curtosis =', curtosis_poblacional)

# Configurar los estilos y paleta de colores de Seaborn
sns.set_style("whitegrid")
sns.set_palette("muted")

# Crear el gráfico combinado de histograma y diagrama de caja
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4), sharex=True)

# Histograma
sns.histplot(x, bins='auto', alpha=0.8, ax=ax1)
ax1.set_xlabel('Valores')
ax1.set_ylabel('Frecuencia')
ax1.set_title('Histograma de los datos')

# Diagrama de caja
sns.boxplot(x=x, orient='h', ax=ax2)
ax2.set_xlabel('Valores')
ax2.set_ylabel('Diagrama de caja')

plt.tight_layout()
plt.show()
```

```python
# LEYENDO DATOS FINANCIEROS DE WEB YAHOO FINANCE ( INSTALAR PREVIAMENTE yfinance)
# pip install yfinance 

#listado de 10 tickers principales (símbolos bursátiles) de empresas reconocidas en Yahoo Finance:
#1.	AAPL – Apple Inc.
#2.	MSFT – Microsoft Corporation
#3.	GOOGL – Alphabet Inc. (Google)
#4.	AMZN – Amazon.com Inc.
#5.	TSLA – Tesla Inc.
#6.	META – Meta Platforms Inc. (Facebook)
#7.	NVDA – NVIDIA Corporation
#8.	JPM – JPMorgan Chase & Co.
#9.	JNJ – Johnson & Johnson
import yfinance as yf
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings

warnings.filterwarnings("ignore")

# --- 1. Descargar los datos de cierre de NVDA Nvidia Corp ---
ticker = yf.Ticker("NVDA")
datos_historicos = ticker.history(period="1mo")  # último mes

# Mostrar resumen del DataFrame
print("Resumen de datos descargados:")
print(datos_historicos[['Open', 'High', 'Low', 'Close', 'Volume']].head())

# --- 2. Convertir columna Close a array ---
x = datos_historicos['Close'].dropna().to_numpy()

# --- 3. Cálculo de estadísticos muestrales ---
media = np.round(np.mean(x), 2)
moda = stats.mode(x)
moda_valor = moda.mode.item()
mediana = np.round(np.median(x), 2)
cuartiles = np.percentile(x, [25, 50, 75])
q1 = np.round(cuartiles[0], 2)
q3 = np.round(cuartiles[2], 2)
amplitud_cuartilica = np.round(q3 - q1, 2)
desv_muestral = np.round(np.std(x, ddof=1), 2)
var_muestral = np.round(np.var(x, ddof=1), 2)
asimetria = np.round(3 * (media - mediana) / desv_muestral, 2)
cv_muestral = np.round((desv_muestral / media) * 100, 2)
curtosis = np.round(stats.kurtosis(x), 2)
valor_max = np.round(np.max(x), 2)
valor_min = np.round(np.min(x), 2)

# --- 4. Cálculo de estadísticos poblacionales ---
desv_poblacional = np.round(np.std(x, ddof=0), 2)
var_poblacional = np.round(np.var(x, ddof=0), 2)
cv_poblacional = np.round((desv_poblacional / media) * 100, 2)

# --- 5. Mostrar resultados ---
print('\n--- Estadísticos Muestrales (NVDA - Cierre) ---\n')
print(f'N° de datos: {len(x)}')
print(f'Valor Máximo: {valor_max}')
print(f'Valor Mínimo: {valor_min}')
print(f'Media: {media}')
print(f'Moda: {moda_valor}')
print(f'Primer Cuartil (Q1): {q1}')
print(f'Mediana (Q2): {mediana}')
print(f'Tercer Cuartil (Q3): {q3}')
print(f'Amplitud Cuartílica: {amplitud_cuartilica}')
print(f'Desviación Estándar Muestral: {desv_muestral}')
print(f'Varianza Muestral: {var_muestral}')
print(f'Coef. de Asimetría de Pearson: {asimetria}')
print(f'Coef. de Variación Muestral (%): {cv_muestral}')
print(f'Curtosis: {curtosis}')

# --- 6. Visualizaciones ---
sns.set_style("whitegrid")
sns.set_palette("muted")

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 6), sharex=False)

# Histograma
sns.histplot(x, bins='auto', alpha=0.8, ax=ax1)
ax1.set_title('Histograma de Cierres de NVDA')
ax1.set_ylabel('Frecuencia')

# Diagrama de caja
sns.boxplot(x=x, orient='h', ax=ax2)
ax2.set_xlabel('Precio de Cierre')
ax2.set_ylabel('')

# Serie temporal del cierre
fechas = datos_historicos.index
cierres = datos_historicos['Close']
ax3.plot(fechas, cierres, marker='o', linestyle='-', color='blue')
ax3.set_title('Serie Temporal de Precios de Cierre (NVDA)')
ax3.set_ylabel('Precio de Cierre')
ax3.set_xlabel('Fecha')
ax3.grid(True)

plt.tight_layout()
plt.show()
```
